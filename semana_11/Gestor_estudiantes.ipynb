````python
import pandas as pd
import numpy as np
from typing import Dict, List, Tuple, Optional
from datetime import datetime
import json
import os

# Funciones auxiliares internas usadas por varias partes del reto
def _calificaciones_con_promedio(df_calificaciones: pd.DataFrame) -> pd.DataFrame:
    """Agrega el promedio por materia usando parcial_1, parcial_2 y final."""
    df = df_calificaciones.copy()
    columnas = ['parcial_1', 'parcial_2', 'final']
    df['promedio_materia'] = df[columnas].mean(axis=1, skipna=True)
    return df

def _metricas_por_estudiante(df_calificaciones: pd.DataFrame) -> pd.DataFrame:
    """Calcula promedio, materias aprobadas y reprobadas por estudiante."""
    df = _calificaciones_con_promedio(df_calificaciones)
    metricas = (
        df.groupby('boleta')
          .agg(
              promedio=('promedio_materia', 'mean'),
              materias_cursadas=('materia_id', 'count'),
              materias_aprobadas=('promedio_materia', lambda x: int((x >= 6).sum())),
              materias_reprobadas=('promedio_materia', lambda x: int((x < 6).sum()))
          )
          .reset_index()
    )
    return metricas
````

````python
def info_general(df_estudiantes: pd.DataFrame, df_calificaciones: pd.DataFrame) -> Dict:
    """
    Genera información general del sistema.
    """
    resultado = {
        "total_estudiantes": int(len(df_estudiantes)),
        "total_registros_calif": int(len(df_calificaciones)),
        "semestres": sorted(df_estudiantes['semestre'].dropna().unique().tolist()),
        "materias_con_registros": int(df_calificaciones['materia_id'].nunique())
    }

    return resultado
````

````python
def validar_datos(df_calificaciones: pd.DataFrame) -> Dict:
    """
    Valida la integridad de los datos.
    """
    columnas_calificaciones = ['parcial_1', 'parcial_2', 'final']

    registros_con_nulos = int(df_calificaciones.isna().any(axis=1).sum())

    fuera_rango = (
        (df_calificaciones[columnas_calificaciones] < 0) |
        (df_calificaciones[columnas_calificaciones] > 10)
    ).any(axis=1)

    calificaciones_fuera_rango = int(fuera_rango.sum())

    resultado = {
        "registros_con_nulos": registros_con_nulos,
        "calificaciones_fuera_rango": calificaciones_fuera_rango,
        "datos_validos": registros_con_nulos == 0 and calificaciones_fuera_rango == 0
    }

    return resultado
````

````python
# PARTE 2: Consultas y Filtros

def buscar_estudiante(df_estudiantes: pd.DataFrame, criterio: str, valor: str) -> pd.DataFrame:
    """
    Busca estudiantes por diferentes criterios.

    Criterios válidos:
    - boleta: búsqueda exacta
    - nombre: búsqueda parcial, sin distinguir mayúsculas/minúsculas
    - semestre: búsqueda exacta convirtiendo el valor a entero
    """
    criterio = criterio.lower().strip()

    if criterio == 'boleta':
        resultado = df_estudiantes[df_estudiantes['boleta'].astype(str) == str(valor)]

    elif criterio == 'nombre':
        resultado = df_estudiantes[
            df_estudiantes['nombre'].astype(str).str.contains(str(valor), case=False, na=False, regex=False)
        ]

    elif criterio == 'semestre':
        try:
            semestre = int(valor)
            resultado = df_estudiantes[df_estudiantes['semestre'] == semestre]
        except ValueError:
            resultado = df_estudiantes.iloc[0:0]

    else:
        raise ValueError("Criterio no válido. Usa: 'boleta', 'nombre' o 'semestre'.")

    return resultado.copy().reset_index(drop=True)
````

````python
def obtener_kardex(boleta: str, df_estudiantes: pd.DataFrame,
                   df_calificaciones: pd.DataFrame, df_materias: pd.DataFrame) -> Dict:
    """
    Obtiene el kardex completo de un estudiante.
    """
    resultado = {
        "estudiante": None,
        "materias": None,
        "promedio_general": 0.0,
        "creditos_cursados": 0,
        "materias_aprobadas": 0,
        "materias_reprobadas": 0
    }

    estudiante_df = df_estudiantes[df_estudiantes['boleta'].astype(str) == str(boleta)]
    if estudiante_df.empty:
        return resultado

    resultado["estudiante"] = estudiante_df.iloc[0].to_dict()

    calif_est = df_calificaciones[df_calificaciones['boleta'].astype(str) == str(boleta)].copy()
    if calif_est.empty:
        resultado["materias"] = pd.DataFrame()
        return resultado

    calif_est = _calificaciones_con_promedio(calif_est)

    materias = calif_est.merge(df_materias, on='materia_id', how='left')
    materias = materias.rename(columns={'nombre': 'materia'})
    materias['estado'] = np.where(materias['promedio_materia'] >= 6, 'Aprobada', 'Reprobada')

    columnas = [
        'materia_id', 'materia', 'creditos',
        'parcial_1', 'parcial_2', 'final', 'promedio_materia', 'estado'
    ]
    materias = materias[columnas].copy()
    materias['promedio_materia'] = materias['promedio_materia'].round(2)

    resultado["materias"] = materias
    resultado["promedio_general"] = round(float(materias['promedio_materia'].mean()), 2)
    resultado["creditos_cursados"] = int(materias['creditos'].fillna(0).sum())
    resultado["materias_aprobadas"] = int((materias['promedio_materia'] >= 6).sum())
    resultado["materias_reprobadas"] = int((materias['promedio_materia'] < 6).sum())

    return resultado
````

````python
def filtrar_por_rendimiento(df_calificaciones: pd.DataFrame,
                            df_estudiantes: pd.DataFrame,
                            min_promedio: float = None,
                            max_promedio: float = None) -> pd.DataFrame:
    """
    Filtra estudiantes por rango de promedio.
    """
    metricas = _metricas_por_estudiante(df_calificaciones)
    resultado = df_estudiantes.merge(metricas[['boleta', 'promedio']], on='boleta', how='left')

    if min_promedio is not None:
        resultado = resultado[resultado['promedio'] >= min_promedio]
    if max_promedio is not None:
        resultado = resultado[resultado['promedio'] <= max_promedio]

    resultado['promedio'] = resultado['promedio'].round(2)
    resultado = resultado.sort_values('promedio', ascending=False)

    return resultado.reset_index(drop=True)
````

````python
# PARTE 3: Cálculos y Estadísticas

def calcular_promedio_materia(df_calificaciones: pd.DataFrame, materia_id: str) -> Dict:
    """
    Calcula estadísticas de una materia.
    """
    calif_materia = df_calificaciones[df_calificaciones['materia_id'] == materia_id].copy()

    resultado = {
        "materia": materia_id,
        "inscritos": int(len(calif_materia)),
        "promedio_parcial1": 0.0,
        "promedio_parcial2": 0.0,
        "promedio_final": 0.0,
        "promedio_general": 0.0,
        "tasa_aprobacion": 0.0,
        "calificacion_maxima": 0.0,
        "calificacion_minima": 0.0
    }

    if calif_materia.empty:
        return resultado

    calif_materia = _calificaciones_con_promedio(calif_materia)

    resultado["promedio_parcial1"] = round(float(calif_materia['parcial_1'].mean()), 2)
    resultado["promedio_parcial2"] = round(float(calif_materia['parcial_2'].mean()), 2)
    resultado["promedio_final"] = round(float(calif_materia['final'].mean()), 2)
    resultado["promedio_general"] = round(float(calif_materia['promedio_materia'].mean()), 2)
    resultado["tasa_aprobacion"] = round(float((calif_materia['promedio_materia'] >= 6).mean() * 100), 1)
    resultado["calificacion_maxima"] = round(float(calif_materia['promedio_materia'].max()), 2)
    resultado["calificacion_minima"] = round(float(calif_materia['promedio_materia'].min()), 2)

    return resultado
````

````python
def ranking_estudiantes(df_calificaciones: pd.DataFrame,
                        df_estudiantes: pd.DataFrame,
                        top_n: int = 10) -> pd.DataFrame:
    """
    Genera ranking de mejores estudiantes por promedio.
    """
    metricas = _metricas_por_estudiante(df_calificaciones)

    ranking = (
        metricas.merge(df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta', how='left')
                .sort_values('promedio', ascending=False)
                .head(top_n)
                .reset_index(drop=True)
    )

    ranking.insert(0, 'Posición', range(1, len(ranking) + 1))
    ranking = ranking.rename(columns={
        'nombre': 'Nombre',
        'semestre': 'Semestre',
        'promedio': 'Promedio'
    })
    ranking['Promedio'] = ranking['Promedio'].round(2)

    return ranking[['Posición', 'Nombre', 'Semestre', 'Promedio']]
````

````python
def estadisticas_por_semestre(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame) -> pd.DataFrame:
    """
    Calcula estadísticas agrupadas por semestre.
    """
    metricas = _metricas_por_estudiante(df_calificaciones)
    por_estudiante = metricas.merge(df_estudiantes[['boleta', 'semestre']], on='boleta', how='left')

    calif = _calificaciones_con_promedio(df_calificaciones)
    calif = calif.merge(df_estudiantes[['boleta', 'semestre']], on='boleta', how='left')

    resumen = (
        por_estudiante.groupby('semestre')
        .agg(
            Estudiantes=('boleta', 'count'),
            Promedio=('promedio', 'mean'),
            **{'Mejor Promedio': ('promedio', 'max'),
               'Peor Promedio': ('promedio', 'min')}
        )
    )

    tasa = calif.groupby('semestre')['promedio_materia'].apply(lambda x: (x >= 6).mean() * 100)
    resumen['Tasa Aprob.'] = tasa

    resumen = resumen[['Estudiantes', 'Promedio', 'Tasa Aprob.', 'Mejor Promedio', 'Peor Promedio']]
    resumen.index.name = 'Semestre'

    for col in ['Promedio', 'Tasa Aprob.', 'Mejor Promedio', 'Peor Promedio']:
        resumen[col] = resumen[col].round(2)

    return resumen
````

````python
# PARTE 4: Identificación de Riesgo y Reportes

def identificar_estudiantes_riesgo(df_calificaciones: pd.DataFrame,
                                   df_estudiantes: pd.DataFrame,
                                   umbral_promedio: float = 7.0,
                                   max_reprobadas: int = 2) -> pd.DataFrame:
    """
    Identifica estudiantes en riesgo académico.

    Criterios:
    - Promedio general menor que umbral_promedio
    - Más de max_reprobadas materias reprobadas
    """
    metricas = _metricas_por_estudiante(df_calificaciones)
    datos = metricas.merge(df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta', how='left')

    bajo_promedio = datos['promedio'] < umbral_promedio
    muchas_reprobadas = datos['materias_reprobadas'] > max_reprobadas

    riesgo = datos[bajo_promedio | muchas_reprobadas].copy()

    def motivo(row):
        bajo = row['promedio'] < umbral_promedio
        reprobadas = row['materias_reprobadas'] > max_reprobadas
        if bajo and reprobadas:
            return 'Ambos'
        if bajo:
            return 'Bajo promedio'
        return 'Mat. reprob.'

    if riesgo.empty:
        return pd.DataFrame(columns=['Boleta', 'Nombre', 'Semestre', 'Promedio', 'Reprobadas', 'Motivo'])

    riesgo['Motivo'] = riesgo.apply(motivo, axis=1)
    riesgo = riesgo.rename(columns={
        'boleta': 'Boleta',
        'nombre': 'Nombre',
        'semestre': 'Semestre',
        'promedio': 'Promedio',
        'materias_reprobadas': 'Reprobadas'
    })
    riesgo['Promedio'] = riesgo['Promedio'].round(2)

    return riesgo[['Boleta', 'Nombre', 'Semestre', 'Promedio', 'Reprobadas', 'Motivo']].reset_index(drop=True)
````

````python
def generar_reporte_academico(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame,
                              df_materias: pd.DataFrame) -> Dict:
    """
    Genera reporte académico completo.
    """
    reporte = {
        "resumen_general": {},
        "por_semestre": None,
        "por_materia": None,
        "mejores_estudiantes": None,
        "estudiantes_riesgo": None,
        "fecha_generacion": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

    calif = _calificaciones_con_promedio(df_calificaciones)
    metricas = _metricas_por_estudiante(df_calificaciones)

    promedio_global = float(metricas['promedio'].mean()) if len(metricas) > 0 else 0.0
    tasa_aprobacion = float((calif['promedio_materia'] >= 6).mean() * 100) if len(calif) > 0 else 0.0

    reporte["resumen_general"] = {
        "total_estudiantes": int(len(df_estudiantes)),
        "total_registros_calif": int(len(df_calificaciones)),
        "total_materias": int(len(df_materias)),
        "promedio_global": round(promedio_global, 2),
        "tasa_aprobacion": round(tasa_aprobacion, 1)
    }

    reporte["por_semestre"] = estadisticas_por_semestre(df_estudiantes, df_calificaciones)

    por_materia = pd.DataFrame([
        calcular_promedio_materia(df_calificaciones, materia_id)
        for materia_id in df_materias['materia_id']
    ])
    por_materia = por_materia.merge(
        df_materias[['materia_id', 'nombre', 'creditos']],
        left_on='materia',
        right_on='materia_id',
        how='left'
    )
    por_materia = por_materia.drop(columns=['materia_id']).rename(columns={'nombre': 'nombre_materia'})
    por_materia = por_materia[
        ['materia', 'nombre_materia', 'creditos', 'inscritos', 'promedio_parcial1',
         'promedio_parcial2', 'promedio_final', 'promedio_general',
         'tasa_aprobacion', 'calificacion_maxima', 'calificacion_minima']
    ]

    reporte["por_materia"] = por_materia
    reporte["mejores_estudiantes"] = ranking_estudiantes(df_calificaciones, df_estudiantes, top_n=10)
    reporte["estudiantes_riesgo"] = identificar_estudiantes_riesgo(df_calificaciones, df_estudiantes)

    return reporte
````

````python
def exportar_kardex(boleta: str, kardex: Dict, formato: str = 'csv') -> str:
    """
    Exporta el kardex de un estudiante a archivo CSV o JSON.
    """
    formato = formato.lower().strip()
    if formato not in ['csv', 'json']:
        raise ValueError("Formato no válido. Usa 'csv' o 'json'.")

    if kardex.get('estudiante') is None:
        raise ValueError("No se puede exportar: el kardex no contiene un estudiante válido.")

    fecha = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"kardex_{boleta}_{fecha}.{formato}"

    if formato == 'csv':
        materias = kardex.get('materias')
        if materias is None:
            materias = pd.DataFrame()

        # Se exportan las materias del kardex porque son la parte tabular principal.
        materias.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')

    else:
        kardex_json = kardex.copy()
        if isinstance(kardex_json.get('materias'), pd.DataFrame):
            kardex_json['materias'] = kardex_json['materias'].to_dict(orient='records')

        with open(nombre_archivo, 'w', encoding='utf-8') as archivo:
            json.dump(kardex_json, archivo, ensure_ascii=False, indent=4)

    return nombre_archivo
````

### 1. Predictor de riesgo

````python
def predecir_riesgo_proximo_semestre(df_calificaciones: pd.DataFrame,
                                     df_estudiantes: pd.DataFrame) -> pd.DataFrame:
    """
    Predice estudiantes que podrían estar en riesgo el próximo semestre
    basándose en tendencias de calificaciones.

    Criterios:
    - Tendencia decreciente en parciales: parcial_1 > parcial_2 > final
    - Calificación final menor que parcial_1
    """
    df = _calificaciones_con_promedio(df_calificaciones)

    df['tendencia_decreciente'] = (
        (df['parcial_1'] > df['parcial_2']) &
        (df['parcial_2'] > df['final'])
    )
    df['final_menor_p1'] = df['final'] < df['parcial_1']
    df['alerta'] = df['tendencia_decreciente'] | df['final_menor_p1']

    resumen = (
        df.groupby('boleta')
          .agg(
              Promedio=('promedio_materia', 'mean'),
              Materias_alerta=('alerta', 'sum'),
              Tendencias_decrecientes=('tendencia_decreciente', 'sum'),
              Final_menor_P1=('final_menor_p1', 'sum')
          )
          .reset_index()
    )

    riesgo = resumen[resumen['Materias_alerta'] > 0].copy()
    riesgo = riesgo.merge(df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta', how='left')

    def motivo(row):
        motivos = []
        if row['Tendencias_decrecientes'] > 0:
            motivos.append('Tendencia decreciente')
        if row['Final_menor_P1'] > 0:
            motivos.append('Final menor que P1')
        return ', '.join(motivos)

    if riesgo.empty:
        return pd.DataFrame(columns=['Boleta', 'Nombre', 'Semestre', 'Promedio', 'Materias_alerta', 'Motivo'])

    riesgo['Motivo'] = riesgo.apply(motivo, axis=1)
    riesgo = riesgo.rename(columns={
        'boleta': 'Boleta',
        'nombre': 'Nombre',
        'semestre': 'Semestre'
    })
    riesgo['Promedio'] = riesgo['Promedio'].round(2)

    return riesgo[['Boleta', 'Nombre', 'Semestre', 'Promedio', 'Materias_alerta', 'Motivo']].reset_index(drop=True)
````

### 2. Comparador de rendimiento

````python
def comparar_estudiantes(boleta1: str, boleta2: str,
                         df_calificaciones: pd.DataFrame,
                         df_estudiantes: pd.DataFrame,
                         df_materias: pd.DataFrame) -> Dict:
    """
    Compara el rendimiento de dos estudiantes.
    """
    kardex1 = obtener_kardex(boleta1, df_estudiantes, df_calificaciones, df_materias)
    kardex2 = obtener_kardex(boleta2, df_estudiantes, df_calificaciones, df_materias)

    if kardex1['estudiante'] is None or kardex2['estudiante'] is None:
        return {
            "error": "Una o ambas boletas no existen.",
            "boleta1_encontrada": kardex1['estudiante'] is not None,
            "boleta2_encontrada": kardex2['estudiante'] is not None
        }

    promedio1 = kardex1['promedio_general']
    promedio2 = kardex2['promedio_general']

    if promedio1 > promedio2:
        mejor = kardex1['estudiante']['nombre']
    elif promedio2 > promedio1:
        mejor = kardex2['estudiante']['nombre']
    else:
        mejor = 'Empate'

    materias1 = kardex1['materias'][['materia_id', 'materia', 'promedio_materia']].rename(
        columns={'promedio_materia': f'promedio_{boleta1}'}
    )
    materias2 = kardex2['materias'][['materia_id', 'promedio_materia']].rename(
        columns={'promedio_materia': f'promedio_{boleta2}'}
    )
    comparacion = materias1.merge(materias2, on='materia_id', how='outer')

    return {
        "estudiante_1": kardex1['estudiante'],
        "estudiante_2": kardex2['estudiante'],
        "promedio_1": promedio1,
        "promedio_2": promedio2,
        "diferencia": round(abs(promedio1 - promedio2), 2),
        "mejor_estudiante": mejor,
        "comparacion_materias": comparacion
    }
````
